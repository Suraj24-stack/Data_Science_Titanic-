**1. Categorical Data **

a. Countplot  

```python
sns.countplot(data=df, x='Survived')
plt.title('Distribution of Survival (0 = No, 1 = Yes)')
plt.show()
df['Survived'].value_counts()
```

1. hue='Pclass': Groups and assigns a distinct color to each passenger class.

2. legend=False: Hides the unnecessary color key to keep the graph clean, since your x-axis is already clearly labeled with the classes (1, 2, and 3).

```python
sns.countplot(data=df, x='Pclass', hue='Pclass', palette='Set1',legend=False)
plt.show()
df['Pclass'].value_counts()
```

```python
sns.countplot(data=df, x='Embarked')
plt.title('Embarked : S= Southampton , C = Cherbourg , Q = Queenstow')
plt.show()
df['Embarked'].value_counts()
```

b. Pie Chart

1. Survived Data on pie Chart

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
plt.show()
```

2. Pclass data on Pie Chart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
plt.show()
```

2. ** Numerical Data **

a. Histogram = Introduced as a primary tool for visualizing the distribution of numerical data.

 bins are used to group continuous data points into specific ranges or intervals.

```python
plt.hist(df['Age'], bins=50)
plt.show()
```

b. Displot

 Introduced as an advanced visualization tool for numerical data that builds upon the concept of a histogram.

KDE stands for Kernel Density Estimation.
Used for analyzing numerical data.

PDF refers to the Probability Density Function.

The Y-axis represents the probability density, while the X-axis represents the variable values (e.g., age).

```python
sns.displot(data=df, x='Age', kde=True)
plt.show()

```

Skew = introduced as a measure of the asymmetry of a probability distribution. It helps you understand whether your numerical data is distributed symmetrically or if it leans toward one side.

```python
df['Age'].skew()
```

c. Boxplot

```python
sns.boxplot(data=df, x='Age')
plt.show()
```
